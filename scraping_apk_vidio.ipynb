```python
from google_play_scraper import Sort, reviews
import pandas as pd

# ambil review aplikasi VIDIO
app_id = "com.vidio.android"
all_reviews = []

for i in range(13):
    result, _ = reviews(
        app_id,
        lang='id',
        country='id',
        sort=Sort.NEWEST,
        count=1000
    )
    all_reviews.extend(result)
    print(f"Batch {i+1} tersimpan, total sekarang: {len(all_reviews)}")

print(f"Total review disimpan: {len(all_reviews)}")

# simpan ke CSV
df = pd.DataFrame(all_reviews)
df.to_csv("vidio_reviews.csv", index=False)
print("File tersimpan sebagai vidio_reviews.csv")

df.head()
```
